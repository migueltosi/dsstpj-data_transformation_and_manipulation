## Main Project

### Data Import

```python
import pandas as pd
```

```python
data = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_hospedagem.json')
data
```

```python
data = pd.json_normalize(data['info_moveis'])
data.head(5)
```

```python
columns = list(data.columns)
columns
```

```python
data = data.explode(columns[3:])
data
```

```python
data.reset_index(inplace = True, drop=True)
data.head()
```

```python
data.info()
```

### Numerical data

```python
import numpy as np
```

```python
data['max_hospedes'] = data['max_hospedes'].astype(np.int64)
```

```python
data.info()
```

```python
numerical_columns = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
data[numerical_columns] = data[numerical_columns].astype(np.int64)
```

```python
data['avaliacao_geral'] = data['avaliacao_geral'].astype(np.float64)
```

```python
data.info()
```

```python
data['preco'] = data['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
data['preco'] = data['preco'].astype(np.float64)
```

```python
data.info()
```

```python
data[['taxa_deposito', 'taxa_limpeza']] = data[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
data[['taxa_deposito', 'taxa_limpeza']] = data[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
data.info()
```

### Textual data

```python
data['descricao_local'] = data['descricao_local'].str.lower()
```

```python
data.head()
```

```python
data['descricao_local'][3169]
```

```python
data['descricao_local'] = data['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
```

```python
data['descricao_local'] = data['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ', regex=True)
```

```python
data['descricao_local'] = data['descricao_local'].str.split()
data.head()
```

```python
data['comodidades'] = data['comodidades'].str.replace('\{|}|\"', '', regex = True)
```

```python
data['comodidades'] = data['comodidades'].str.split(',')
data.head()
```

```python
data.head()
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].str.lower()
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True).replace('(?<!\w)-(?!\w)',' ', regex=True)
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].str.split()
```
